```python
#Changing the purpose column to numbers for sklearn to understand
```

```python
catg_feats = ['purpose']
```

```python
clean_data = pd.get_dummies(df, columns=catg_feats, drop_first=True)
```

```python
clean_data
```

```python
df['purpose'].value_counts()
```

```python
#Time to do train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = clean_data.drop('not.fully.paid', axis=1)
y = clean_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
#Using Forest Tree
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 200)
```

```python
rfc
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_preds = rfc.predict(X_test)
```

```python
rfc_preds
```

```python
print(classification_report(y_test, rfc_preds))
print(confusion_matrix(y_test, rfc_preds))
```

```python
# The Random Forest out-performed the Decision Tree model.
```
